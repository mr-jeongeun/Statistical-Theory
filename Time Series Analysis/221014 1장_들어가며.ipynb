```r
a = ts(1:10, frequency = 4, start = c(1959, 2))
class(a)
a
```

```r
z = ts(matrix(rnorm(300), 100, 3), start = c(1961, 1), frequency = 12)
class(z)
head(z)
z
```

```r
plot(z)
plot(z, plot.type = "single", lty = 1:3)
```

## 1.3 시계열 자료의 시각화

```r
AirPassengers
```

```r
AirPassengers.q = aggregate(AirPassengers, nfreq = 4, FUN = sum)
AirPassengers.q
```

```r
AirPassengers.y = aggregate(AirPassengers, nfreq = 1, FUN = sum)
AirPassengers.y
```

```r
data("AirPassengers")
plot.ts(AirPassengers)
ts.plot(AirPassengers)
```

```r
install.packages("tseries")
```

```r
library(tseries)
data(nino)
str(nino)
plot.ts(nino)
ts.plot(nino, col = 1:2)
```

```r
attach(iris)
t.iris = as.ts(iris[, -5])
plot.ts(t.iris)
ts.plot(t.iris, col = 1:4)
```

## 1.4 R 시계열 자료 다루기

```r
x = cumsum(cumsum(1:10))
x
```

```r
diff(x, lag = 2)
```

```r
diff(x, differences = 2)
```

```r
diff(ts(1:10), lag = 2)
```

```r
diff(ts(1:10), lag = 2, differences = 2)
```

```r
x = strptime(c("2006/01/08 10:07:52", "2006/08/07 19:33:02"),
             "%Y/%m/%d %H:%M:%S", tz = "GMT")
str(x)
y = as.POSIXct(x)
str(y)
```
