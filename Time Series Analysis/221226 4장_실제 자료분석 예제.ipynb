## 예제 1. 분기별 자료

### (a) 기초분석

```r
data("UKgas")
class(UKgas)
start(UKgas)
end(UKgas)
frequency(UKgas)
time(UKgas)
```

```r
subset = window(UKgas, start = c(1960, 1), end = c(1970, 3), frequency = 4) 
ts.intersect(UKgas, subset) # frequency = 4, 분기별
ts.union(UKgas, subset)
ts.plot(UKgas)
hist(UKgas)
```

```r
plot(aggregate(UKgas))
cycle(UKgas)
boxplot(UKgas ~ cycle(UKgas))  # 분기에 따른 UKgas
```

```r
## 시계열 분해
d = decompose(UKgas)
plot(d)
sd = stl(UKgas, s.window = "periodic")
plot(sd)
```

```r
trend = d$trend
season = d$seasonal
ts.plot(cbind(trend, trend * season), lty = 1:2)
legend("topleft", c("trend", "trend*seasonality"), lty = 1:2)
monthplot(UKgas)  # 분기에 따른 UKgas
```

```r
install.packages("forecast")
library(forecast)
seasonplot(UKgas)
```

### (b) ARIMA 모형 적합

```r
ndiffs(UKgas)
nsdiffs(UKgas)
```

```r
a = diff(UKgas)
b = diff(a, 4)
acf(b)
pacf(b)
```

```r
fit1 = arima(UKgas, order = c(0, 1, 1), seasonal = list(order = c(0, 1, 0)))
fit1
```

### (c) auto.arima() 함수를 이용한 모형 적합

```r
fit2 = auto.arima(UKgas, trace = TRUE, ic = "aicc")
```

### (d) 잔차분석

```r
par(mfrow = c(2, 1))
acf(residuals(fit2))
pacf(residuals(fit2))
par(mfrow = c(1, 1))
```

### (e) 모형의 성능평가와 예측

```r
library(forecast)
accuracy(fit2)
forecast(fit2, 8)
plot(forecast(fit2, 8))
```

## 예제 2. 월별 자료

### (a) 기초분석

```r
data("USAccDeaths")
class(USAccDeaths)
start(USAccDeaths)
end(USAccDeaths)
frequency(USAccDeaths)
time(USAccDeaths)
ts.plot(USAccDeaths)
hist(USAccDeaths)
```

```r
plot(aggregate(USAccDeaths))
cycle(USAccDeaths)
boxplot(USAccDeaths ~ cycle(USAccDeaths))
```

```r
## 시계열 분해
d = decompose(USAccDeaths)
plot(d)
sd = stl(USAccDeaths, s.window = "period")
plot(sd)
```

```r
trend = d$trend
season = d$seasonal
ts.plot(cbind(trend, trend*season), lty = 1:2)
monthplot(USAccDeaths)
```

```r
library(forecast)
seasonplot(USAccDeaths)
```

### (b) ARIMA 모형 적합

```r
ndiffs(USAccDeaths, test = "adf")
nsdiffs(USAccDeaths)
```

```r
a = diff(USAccDeaths, 12)
par(mfrow = c(1, 2))
acf(a)
pacf(a)
```

```r
fit1 = auto.arima(USAccDeaths, d = 0, D = 1, trace = TRUE, ic = "aicc")
fit2 = auto.arima(USAccDeaths, trace = TRUE, ic = "aicc")
```

### (c) 잔차분석

```r
acf(residuals(fit1))
pacf(residuals(fit1))
```

```r
acf(residuals(fit2))
pacf(residuals(fit2))
```

### (d) 모형의 성능평가와 예측

```r
library(forecast)
fit1$aic
fit2$aic
```

```r
round(accuracy(fit1), 3)
round(accuracy(fit2), 3)
```

```r
par(mfrow = c(1, 1))
plot(forecast(fit2, 5))
plot(forecast(fit2, 12))
```
