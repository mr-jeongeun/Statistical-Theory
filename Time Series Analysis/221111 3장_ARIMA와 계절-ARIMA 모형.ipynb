## 3.6 시계열 모의실험

### (a) 백색잡음 계열

### 예제 1

```r
set.seed(200)
GWN = rnorm(n = 100, mean = 5, sd = 0.2)
PWN = rpois(n = 50, lambda = 10)
```

```r
par(mfrow = c(1, 3))
plot.ts(GWN)
abline(h = 5, col = "blue", lty = "dashed")
acf(GWN, main = "", lag.max = 20)
pacf(GWN, main = "", lag.max = 20)
```

```r
par(mfrow = c(1, 3))
plot.ts(PWN)
abline(h = 10, col = "blue", lty = "dashed")
acf(PWN, main = "", lag.max = 20)
pacf(PWN, main = "", lag.max = 20)
```

### (b) 확률보행 계열

```r
set.seed(200)
n = 100
y = w = rnorm(n, 0, 1)
for(t in 2:n){
  y[t] = y[t-1] + w[t]
}

par(mfrow = c(1, 3))
plot.ts(y, ylab = expression(italic(y[t])))
acf(y)
pacf(y)
```

```r
y1 = cumsum(w)
par(mfrow = c(1, 3))
plot.ts(y1, ylab = expression(italic(y[t])))
acf(y1)
pacf(y1)
```

### (c) 상수항을 가지는 확률보행 계열

```r
par(mfrow = c(1, 1))
set.seed(100)
w = rnorm(100, 0, 1)
x = cumsum(w)
rw = w + 0.5
RW = cumsum(rw)
plot.ts(RW, main = "Random Walk", col = "red", ylim = c(-10, 60))
lines(x)
lines(.5 * (1:200), lty = "dashed")    
```

### (g) 계절-ARIMA 모형

```r
# install.packages("sarima")
```

```r
set.seed(200)
library(sarima)
n = 10000
x = sim_sarima(n, model = list(sar = 0.8, nseasons = 12, sigma2 = 1))

par(mfrow = c(1, 3))
plot.ts(x[1:50], ylab ="")
acf(x, lag = 60, main = "")
pacf(x, lag = 60, main = "")
```
