#### 인덱싱

- R에서 string[1] = H : 인덱싱이 1부터 n까지
- Py에서 string[1] = e : 인덱싱이 0부터 n-1까지

```python
string = 'Hello World'
```

```python
string[0]
```

#### 슬라이싱

```python
print(string[0:4])  # 4-1(= 3)번째까지 추출
print(string[0:5])  # 5-1(= 4)번째까지 추출
```

```python
len(string)  # 문자열 총 길이
```
