## 클래스(class)

위 내용을 우리가 앞에서 익힌 함수를 이용해 구현해 보자. 계산의 "더하기" 기능을 구현한 파이썬 코드는 다음과 같다.

```python
result = 0

def add(num):
  global result
  result += num
  return result

print(add(3))
print(add(4))
print(add(5))
```

그런데 만일 한 프로그램에서 2대의 계산기가 필요한 상황이 발생하면 어떻게 해야 할까? 

각 계산기는 각각의 결괏값을 유지해야 하기 때문에 위와 같이 add 함수 하나만으로는 결괏값을 따로 유지할 수 없다.


```python
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

위와 같은 경우에 클래스를 사용하면 다음과 같이 간단하게 해결할 수 있다.

```python
class Calculator:
  def __init__(self):
    self.result = 0

  def add(self, num):
    self.result += num
    return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(3))
print(cal2.add(7))
```

## Class의 용어 정리

-  부모 클래스(Super Class)
  - 모태가 되는 클래스
- 자식 클래스(Sub Class, Child Class)
  - 부모로 부터 상속 받은 클래스
- 생성자(Constructor)
  - 클래스 인스턴스를 생성
- __init__() 메서드
  - 클래스 생성자 메서드
- 클래스 변수
  - 클래스에서 공유 되는 변수
- 메서드(Method)
- 메서드 오버라이딩(Method Overriding)
  - 상속 받은 메서드를 재정의

### class 생성

```python
class Person:
  total_count = 0  # 클래스 변수

  def __init__(self):   # 생성자 메서드
    self.name = "홍길동"
    self.age = 3
    Person.total_count += 1

  def introduce(self):  # 메서드
    print("제 이름은 {}이고 나이는 {}살입니다.".format(self.name, self.age))
```

```python
Per1 = Person()
Per1.introduce()
```

```python
##주의## 아래의 경우는 self 객체가 선언되지 않아 오류 발생 
Person.introduce()
```

```python
# 부모클래스 Person + 자식클래스 Student
# super() : 부모클래스에 있는 내용을 상속받기 위한 함수
class Student(Person):
  """Sub Class"""
  def __init__(self):
    super().__init__()
```

```python
stu = Student()
stu.introduce()  # Person에서 상속을 받아서 Student는 introduce 선언 없이도 사용 가능
```

```python
p2 = Person()
p2.introduce()
```

```python
p3 = Person()
p3.introduce()
```

```python
print(Per1.__class__.total_count)  # 부모클래스에서 1번 + 자식클래스에서 3번
```

```python
class Person:
  """super class"""
  total_count = 0

  def __init__(self, name, age):    # 생성자 메서드
    self.name = name
    self.age = age
    Person.total_count += 1

  def introduce(self):    # 메서드
    print("제 이름은 {}이고 나이는 {}살 입니다.".format(self.name, self.age))
```

### 상속

```python
class Person:
  """super class"""
  total_count = 0

  def __init__(self, name, age):    # 생성자 메서드
    self.name = name
    self.age = age
    Person.total_count += 1

  def introduce(self):    # 메서드
    print("제 이름은 {}이고 나이는 {}살 입니다.".format(self.name, self.age))

class Student(Person):
  """sub class"""

  def __init__(self):
    super().__init__()
```

```python
stu1 = Student("홍길동", 5)
```

```python
stu2 = Student()
stu2.introduce()
```

### 클래스 매서드 오버라이딩

```python
class Person:
  """super class"""
  total_count = 0

  def __init__(self, name = "홍길동", age = 20):    # 생성자 메서드
    self.name = name
    self.age = age
    Person.total_count += 1

  def introduce(self):    # 메서드
    print("제 이름은 {}이고 나이는 {}살 입니다.".format(self.name, self.age))

class Student(Person):
  """sub class"""

  def __init__(self):
    super().__init__()

  def introduce(self):
    super().introduce()  # 주석 이용 시 자식 class의 내용만 호출
    print(f"난 Student야 내 이름은 {self.name}이고, 나이는 비밀이야")
```

```python
s1 = Student()
s1.introduce()
```
