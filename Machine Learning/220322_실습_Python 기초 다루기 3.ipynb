## 딕셔너리(dictionary)


딕셔너리는 리스트와 비슷하다. 다른 점은 항목의 순서를 따지지 않으며, 0 또는 1과 같은 오프셋으로 항목을 선택할 수 없다.\
대신 **값(value)**에 상응하는 고유한 **키(key)**를 지정한다. 이러한 키는 여러가지 타입이 될 수 있다.

　

- 생성하기:{}
\
　　　딕셔너리를 만들려면 중괄호{} 안에 콤마로 구분한 키:값 쌍을 지정한다.





```python
empty_dict ={}
empty_dict
```

```python
example_dict = {'name':'홍길동','age':'20','phone':'01012345678'}
example_dict
```

- 생성하기 : dict()

　　　따옴표를 많이 입력하는 것을 좋아하지 않는 경우 다음과 같이 딕셔너리를 생성할 수 있다.


```python
example_dict_2 = dict(name = '홍길동',age = '20',phone = '01012345678')
example_dict_2
```

- 항목 추가/변경:[key]

　　　딕셔너리에 항목을 추가하는 것은 간단하다. 키에 참조되는 항목에 값을 할당하면 된다.\
　　　딕셔너리에 이미 존재하는 키라면 그 값은 새 값으로 대체된다.

다음은 나라이름을 key로, 이름을 value로 하는 딕셔너리를 만든 예제이다.\
key에 미국을 value에는 Washington D.C를 추가해 보자.

```python
capital = {
    '한국' : 'Seoul',
    '일본' : 'Tokyo',
    '중국' : 'China'
}

capital
```

```python
capital['미국'] = 'Washington D.C'
```

```python
capital
```

중국의 수도를 China로 잘못 입력 하였다. Beijing으로 수정해 보자.

이미 중국이라는 key가 존재하기 때문에 중국 key에 Beijing 을 입력하면 대체된다.

```python
capital['중국']  = 'Beijing'
```

```python
capital
```

## 딕셔너리의 함수들

- 항목 얻기: [key]와 get() 

딕셔너리에 키를 지정하여 상응하는 값을 얻는다.

```python
capital['미국']
```

다음과 같이 get() 함수를 사용해도 키에 상응하는 값을 얻을 수 있다.

```python
capital.get('미국')
```

키가 존재하지 않으면, 옵션값을 지정해서 이를 출력할 수 있다.

```python
capital.get('캐나다','Not a Python')
```

옵션값을 지정하지 않으면 None을 얻는다. 대화식 인터프리터는 아무것도 출력하지 않는다.

```python
capital.get('캐나다')
```

- 모든 키 얻기 : keys()

```python
capital.keys()
```



> keys()는 파이썬 ver.2 에서는 리스트를 반환하고, 파이썬 ver. 3 에서는 dict_keys()를 반환한다.



- 모든 값 얻기 : values()

딕셔너리의 모든 값을 가져오기 위해 values()를 사용한다.

```python
capital.values()
# list로 변환 list(capital.values())
```

- 딕셔너리의 모든 키/값을 얻기 위해 items()를 사용한다. **(매우 많이 쓰는 함수)**

```python
list(capital.items())
```

## 셋(Set)

**set**은 값은 버리고 키만 남은 딕셔너리와 같다. 보통 **list**나 **numpy** 만으로도 해결 가능해서 **set은** 평소에 잘 사용 하지않는다. 하지만 딕셔너리와 마찬가지로 각 키는 유일하기 때문에 리스트에 어떤 것이 존재하는지 판단할 때 사용된다. 다음 예를 통해서 알아보자

```python
example_set = ['a','a','a','b','b','c','c']
set(example_set)
```
