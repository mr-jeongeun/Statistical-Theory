#### 배열(array)

#### if문

```python
score = int(input("점수를 입력해주세요(1 ~ 100) : "))
```

```python
if score >= 90:
  print('A등급')
elif score >= 80 and score < 90:
  print('B등급')
elif score >= 70 and score < 80:
  print('C등급')
else:
  print('D등급')
```

#### for문

#### 범위 설정
- 1 : 10
- range(11)
- range(0, 11)

```python
cnt = 0
for i in range(11):
  cnt += i
print(cnt)
```

#### while문

```python
cnt = 0
i = 1
while i < 11:
  cnt += i
  i = i + 1
print(cnt)
```
