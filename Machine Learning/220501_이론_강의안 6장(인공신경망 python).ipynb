### 1. 범주형 데이터

```python
# 데이터 가져오기"
import pandas as pd
mydata = pd.read_csv('https://stats.idre.ucla.edu/stat/data/binary.csv')
mydata
```

```python
# 데이터 분할
X = mydata.iloc[:, 1:]
y = mydata.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)  # 평가데이터 25%
```

```python
# 데이터 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
# nnet 적합
from sklearn.neural_network import MLPClassifier
clf2 = MLPClassifier(hidden_layer_sizes = 2, random_state = 1)
print(clf2.fit(X, y))
clf4 = MLPClassifier(hidden_layer_sizes = 4, random_state = 1)
print(clf4.fit(X, y))
clf6 = MLPClassifier(hidden_layer_sizes = 6, random_state = 1)
print(clf6.fit(X, y))
clf10 = MLPClassifier(hidden_layer_sizes = 10, random_state = 1)
print(clf10.fit(X, y))
```

```python
print("clf2 학습 데이터 점수 : {}".format(clf2.score(X_train,y_train)))
print("clf2 평가 데이터 점수 : {}".format(clf2.score(X_test,y_test)))

from sklearn.metrics import accuracy_score
predict = clf2.predict(X_test)
print("clf2 정확도 : {}".format(accuracy_score(y_test,predict)))

from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_true = y_test, y_pred = predict)
print(confmat)
```

```python
print("clf4 학습 데이터 점수 : {}".format(clf4.score(X_train,y_train)))
print("clf4 평가 데이터 점수 : {}".format(clf4.score(X_test,y_test)))

from sklearn.metrics import accuracy_score
predict = clf4.predict(X_test)
print("clf4 정확도 : {}".format(accuracy_score(y_test,predict)))

from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_true = y_test, y_pred = predict)
print(confmat)
```

```python
print("clf6 학습 데이터 점수 : {}".format(clf6.score(X_train,y_train)))
print("clf6 평가 데이터 점수 : {}".format(clf6.score(X_test,y_test)))

from sklearn.metrics import accuracy_score
predict = clf6.predict(X_test)
print("clf6 정확도 : {}".format(accuracy_score(y_test,predict)))

from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_true = y_test, y_pred = predict)
print(confmat)
```

```python
print("clf10 학습 데이터 점수 : {}".format(clf10.score(X_train,y_train)))
print("clf10 평가 데이터 점수 : {}".format(clf10.score(X_test,y_test)))

from sklearn.metrics import accuracy_score
predict = clf10.predict(X_test)
print("clf10 정확도 : {}".format(accuracy_score(y_test,predict)))

from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_true = y_test, y_pred = predict)
print(confmat)
```
